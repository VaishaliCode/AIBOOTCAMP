```python


import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Day": ["Day 1", "Day 2", "Day 3", "Day 4", "Day 5"],
    "Visitors": [5000, 7000, 8500, 9200, 11000]
}
```

```python
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df["Day"], df["Visitors"], marker='o', linestyle='-', color='b', label="Visitors")
plt.xlabel("Days")
plt.ylabel("Number of Visitors")
plt.title("Visitors Over Days at Khubh Mela")
plt.grid(True)
plt.legend()
plt.savefig("visitors_line_plot.png")  # Save the plot
plt.show()
```

```python
city_data = {
    "City": ["Mumbai", "Delhi", "Pune", "Bangalore", "Kolkata"],
    "Participants": [1500, 2300, 1800, 2100, 1700]
}
```

```python
df_city = pd.DataFrame(city_data)
```

```python
plt.figure(figsize=(8, 5))
plt.bar(df_city["City"], df_city["Participants"], color="lightgreen", label="Participants")
plt.xlabel("City")
plt.ylabel("Number of Participants")
plt.title("Participants from Different Cities")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.legend()
plt.savefig("participants_bar_chart.png")  # Save the plot
plt.show()
```
